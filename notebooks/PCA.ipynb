```python
# Count of many of each cell type in the dataframe (columns)

cell_type_counts = df.columns[1:].str.split(".", n=1).str[0].value_counts()
print(cell_type_counts)

cell_type_counts.sum() == (df.shape[1] - 1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# Preprocessing for PCA
df_beta = df.drop(columns="probe_ID") # we drop probe_ID column
print(df_beta.shape) # check dimensions. Standard format for sklearn: row/line = sample/observation, column = variable/feature. OK!

if df_beta.isna().any().any():
    warnings.warn(
        "Missing values detected in beta matrix. "
        "Filling NaNs with column means before PCA.",
        RuntimeWarning
    )
    df_beta = df_beta.fillna(df_beta.mean())
else:
    print("No missing values detected in beta matrix.")


scaler = StandardScaler(with_std=False) # We center data -> every tissue gets mean = 0. We do not want to scale, we want to preserve biological information and variance.

X_centered = scaler.fit_transform(df_beta.values)

# Run PCA
pca = PCA(
    n_components=20,
    svd_solver="randomized",
    random_state=42
)

X_pca = pca.fit_transform(X_centered)

print("Shape of PCA:", X_pca.shape)

print("Explained variance:", pca.explained_variance_ratio_)
print("Cumulative:", np.cumsum(pca.explained_variance_ratio_))



# Hmm what do we want to reduce? is it columns or is it rows?
# I guess we want to cluster with all datapoints -> Columns reduce, how much?
```
